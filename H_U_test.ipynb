```python
# %% Import required libraries
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import joblib

# %% Load dataset
data_sup = pd.read_csv('C:/Users/ayama/OneDrive/Bureau/diet_system/data/cleaned_recipes_.csv')
print("Dataset Loaded Successfully")
```

```python
# %% Extract features and labels
Ingredients = data_sup['RecipeIngredientParts']
labels = data_sup['HealthStatus']
```

```python
from sklearn.utils import resample

# Resample X and labels to 30% of the original dataset
sample_size = int(0.5 * len(data_sup))  # 30% of the dataset size

# Resample X and labels
#X_resampled, y_resampled = resample(Ingredients, labels, n_samples=sample_size, random_state=42)

# Initialize the CountVectorizer
vectorizer = CountVectorizer(tokenizer=lambda x: x.split(", "), lowercase=True)

# Transform the ingredients into a sparse matrix
X_sparse = vectorizer.fit_transform(Ingredients)

# Check the shape and type of the sparse matrix
print("Shape of sparse matrix:", X_sparse.shape)
```

```python

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_sparse, labels, test_size=0.2, random_state=42)

```

```python
from sklearn.linear_model import LogisticRegression

# Initialize and train the Logistic Regression model with regularization (C controls the strength)
model_lr_ = LogisticRegression(C=1.0)  # Decrease C for stronger regularization
model_lr_.fit(X_train, y_train)
```

```python

# Make predictions on the test set
y_pred = model_lr_.predict(X_test)

# Evaluate the model
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

random forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

# Train a RandomForestClassifier
model_rf = RandomForestClassifier()
model_rf.fit(X_train, y_train)
```

```python

# Make predictions on the test set
y_pred = model_rf.predict(X_test)
cm = classification_report(y_test, y_pred)

report = confusion_matrix(y_test, y_pred)
# Evaluate the model
print("Classification Report:\n", cm)
print("Confusion Matrix:\n", report)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Plot the confusion matrix using seaborn heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Unhealthy', 'Healthy'], yticklabels=['Unhealthy', 'Healthy'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

print ('__________________________________________________________________')


# Convert the report into a pandas DataFrame for easier plotting
report_df = pd.DataFrame(report).transpose()

# Plot Precision, Recall, and F1-score for each class
metrics = ['precision', 'recall', 'f1-score']
plt.figure(figsize=(10, 6))

for i, metric in enumerate(metrics):
    plt.subplot(1, 3, i + 1)
    report_df[metric].drop('accuracy').plot(kind='bar', color='skyblue', figsize=(10, 6))
    plt.title(f'{metric.capitalize()} by Class')
    plt.ylabel(metric.capitalize())
    plt.xlabel('Class')
    plt.xticks(rotation=45)
    plt.tight_layout()

plt.show()
```

```python

# Function to process user input and predict health status
def predict_health_status(user_input):
    # Preprocess user input (same as training data)
    user_input_processed = vectorizer.transform([user_input])
    
    # Predict health status (0 = Unhealthy, 1 = Healthy)
    prediction = model_rf.predict(user_input_processed)

    # Find nearest neighbors
    distances, indices = model_nn.kneighbors(user_input_processed, n_neighbors=10)
    
    # Retrieve similar recipes
    similar_recipes = data_sup.iloc[indices[0]].copy()
    similar_recipes['Distance'] = distances[0]
    
     
    # Return the result
    return prediction , similar_recipes
    

# Test with user input
user_input = input("Enter ingredients (comma separated): ")
result , similar_recipes = predict_health_status(user_input)

print(f"The health status of the ingredients is: {result}")
print(f"recipes as your ingrediants \n")
similar_recipes
```
